# SISSO Equation Post-Processing Tool

**PURPOSE**: This notebook evaluates SISSO symbolic regression equations to:
1. Calculate predictions on material property datasets
2. Assess model performance with statistical metrics (R², RMSE, MAE)
3. Generate publication-quality parity plots
4. Export detailed results and visualizations

**REQUIREMENTS**:
* Python 3.12+
* Libraries: pandas (2.2.3), matplotlib (3.9.2), numpy (1.26.4), scikit-learn (1.5.2), colorama (0.4.6), openpyxl
* Input file: `SISSO_post-processing_input.xlsx` with property sheets

**USAGE**:
1. Run the notebook cells sequentially
2. Follow the interactive prompts to:
   * Load your Excel file and select property sheet
   * Choose formula, actual values, and parameter columns
   * Enter SISSO equation with coefficients
   * Validate equation on sample data
   * Specify output location and filenames

**NOTE**: Use `**` for exponentiation (not `^`) in equations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from matplotlib.ticker import MaxNLocator, AutoMinorLocator, ScalarFormatter
import os
from pathlib import Path
import re
from colorama import init, Fore, Style, Back

# Initialize colorama for cross-platform colored output
init(autoreset=True)

plt.rcParams['font.family'] = 'sans-serif'


# ================================
# UTILITY FUNCTIONS
# ================================

def print_header(text):
    """Print a colored header"""
    print(f"\n{Back.BLUE}{Fore.WHITE}{Style.BRIGHT} {text} {Style.RESET_ALL}\n")


def print_success(text):
    """Print success message in magenta"""
    print(f"{Fore.MAGENTA}[SUCCESS] {text}{Style.RESET_ALL}")


def print_error(text):
    """Print error message in red"""
    print(f"{Fore.RED}[ERROR] {text}{Style.RESET_ALL}")


def print_warning(text):
    """Print warning message in yellow"""
    print(f"{Fore.YELLOW}[WARNING] {text}{Style.RESET_ALL}")


def print_info(text):
    """Print info message in green"""
    print(f"{Fore.GREEN}[INFO] {text}{Style.RESET_ALL}")


def get_valid_path(prompt_text, must_exist=True):
    """Get a valid file/directory path from user with error handling"""
    while True:
        path_input = input(f"{Fore.MAGENTA}{prompt_text}{Style.RESET_ALL}").strip()
        path_input = path_input.strip('"').strip("'")  # Remove quotes if present
        
        if not path_input:
            print_error("Path cannot be empty. Please try again.")
            continue
        
        path_obj = Path(path_input)
        
        if must_exist:
            if path_obj.exists():
                print_success(f"Path found: {path_obj}")
                return str(path_obj)
            else:
                print_error(f"Path does not exist: {path_obj}")
                retry = input(f"{Fore.YELLOW}Try again? (y/n): {Style.RESET_ALL}").lower()
                if retry != 'y':
                    return None
        else:
            return str(path_obj)


def get_valid_string(prompt_text, allow_empty=False):
    """Get a valid string input from user"""
    while True:
        user_input = input(f"{Fore.MAGENTA}{prompt_text}{Style.RESET_ALL}").strip()
        
        if not user_input and not allow_empty:
            print_error("Input cannot be empty. Please try again.")
            continue
        
        return user_input


def get_valid_float(prompt_text):
    """Get a valid float input from user with error handling"""
    while True:
        user_input = input(f"{Fore.MAGENTA}{prompt_text}{Style.RESET_ALL}").strip()
        
        try:
            value = float(user_input)
            return value
        except ValueError:
            print_error(f"'{user_input}' is not a valid number. Please enter a numeric value.")


def get_valid_int(prompt_text, min_val=None, max_val=None):
    """Get a valid integer input from user with error handling"""
    while True:
        user_input = input(f"{Fore.CYAN}{prompt_text}{Style.RESET_ALL}").strip()
        
        try:
            value = int(user_input)
            if min_val is not None and value < min_val:
                print_error(f"Value must be at least {min_val}.")
                continue
            if max_val is not None and value > max_val:
                print_error(f"Value must be at most {max_val}.")
                continue
            return value
        except ValueError:
            print_error(f"'{user_input}' is not a valid integer. Please enter a whole number.")


def select_from_list(items, prompt_text):
    """Allow user to select an item from a list"""
    print(f"\n{Fore.YELLOW}{prompt_text}{Style.RESET_ALL}")
    for i, item in enumerate(items, 1):
        print(f"  {Fore.CYAN}{i}.{Style.RESET_ALL} {item}")
    
    while True:
        choice = get_valid_int(f"Enter choice (1-{len(items)}): ", min_val=1, max_val=len(items))
        return items[choice - 1]


# ================================
# MAIN FUNCTIONS
# ================================

def load_data():
    """Load Excel data with user inputs"""
    print_header("DATA LOADING")
    
    # Get Excel file directory
    print_info("Step 1: Specify the folder containing your Excel file (use './' if in current folder)")
    file_dir = get_valid_path("Enter the folder path: ", must_exist=True)
    if not file_dir:
        return None
    
    # List available Excel files in the directory
    excel_files = list(Path(file_dir).glob('*.xlsx')) + list(Path(file_dir).glob('*.xls'))
    
    if excel_files:
        print_success(f"Found {len(excel_files)} Excel file(s) in the directory:")
        for i, file in enumerate(excel_files, 1):
            print(f"  {i}. {file.name}")
    else:
        print_warning("No Excel files found in this directory")
    
    # Get Excel file name
    print_info("\n Step 2: Specify the Excel file name")
    file_name = get_valid_string("Enter the Excel file name (with extension .xlsx or .xls): ")
    
    # Ensure file has proper extension
    if not file_name.endswith('.xlsx') and not file_name.endswith('.xls'):
        print_warning("Adding .xlsx extension to filename")
        file_name += '.xlsx'
    
    # Construct full path
    excel_path = str(Path(file_dir) / file_name)
    
    # Check if file exists
    if not Path(excel_path).exists():
        print_error(f"File not found: {excel_path}")
        retry = input(f"{Fore.YELLOW}Try again? (y/n): {Style.RESET_ALL}").lower()
        if retry == 'y':
            return load_data()
        return None
    
    print_success(f"File located: {excel_path}")
    
    # Load Excel file to get sheet names
    try:
        xls = pd.ExcelFile(excel_path)
        sheet_names = xls.sheet_names
        print_success(f"Found {len(sheet_names)} sheet(s) in the Excel file")
    except Exception as e:
        print_error(f"Error reading Excel file: {e}")
        return None
    
    # Select sheet
    sheet_name = select_from_list(sheet_names, "Select the sheet to analyze:")
    print_success(f"Selected sheet: {sheet_name}")
    
    # Load the selected sheet
    try:
        df = pd.read_excel(excel_path, sheet_name=sheet_name)
        print_success(f"Loaded {len(df)} rows and {len(df.columns)} columns")
        print_info("Available columns:")
        for i, col in enumerate(df.columns, 1):
            print(f"  {Fore.MAGENTA}{i}.{Style.RESET_ALL} {col}")
    except Exception as e:
        print_error(f"Error loading sheet: {e}")
        return None
    
    return excel_path, sheet_name, df


def select_columns(df):
    """Select columns for analysis"""
    print_header("COLUMN SELECTION")
    
    columns = list(df.columns)
    
    # Select formula column
    formula_col = select_from_list(columns, "Select the CHEMICAL FORMULA column:")
    print_success(f"Formula column: {formula_col}")
    
    # Select actual values column
    actual_col = select_from_list(columns, "Select the ACTUAL VALUES column:")
    print_success(f"Actual values column: {actual_col}")
    
    # Select parameter columns for SISSO
    print(f"\n{Fore.YELLOW} Select PARAMETER columns for SISSO prediction{Style.RESET_ALL}")
    print_info("These are the columns containing PLSR selected parameters")
    print("\n Available columns:")
    for i, col in enumerate(columns, 1):
        if col not in [formula_col, actual_col]:
            print(f"  {Fore.GREEN}{i}.{Style.RESET_ALL} {col}")
    
    print_info("Enter column numbers separated by commas (e.g., 3,4,5)")
    
    while True:
        param_input = input(f"{Fore.CYAN}Enter column numbers: {Style.RESET_ALL}").strip()
        try:
            param_indices = [int(x.strip()) - 1 for x in param_input.split(',')]
            param_cols = [columns[i] for i in param_indices if 0 <= i < len(columns)]
            
            if not param_cols:
                print_error("No valid columns selected. Please try again.")
                continue
            
            print_success(f"Selected {len(param_cols)} parameter columns:")
            for col in param_cols:
                print(f"  • {col}")
            
            confirm = input(f"{Fore.YELLOW}Confirm selection? (y/n): {Style.RESET_ALL}").lower()
            if confirm == 'y':
                return formula_col, actual_col, param_cols
        except (ValueError, IndexError):
            print_error("Invalid input. Please enter valid column numbers separated by commas.")


def create_column_aliases(param_cols):
    """
    Create simplified aliases for column names with spaces or special characters
    """
    aliases = {}
    reverse_aliases = {}
    
    for col in param_cols:
        # Check if column name has spaces or special characters
        if ' ' in col or any(c in col for c in ['(', ')', '-', '+', '*', '/', '^']):
            # Create a simplified alias
            # Remove special characters and spaces, convert to lowercase
            alias = re.sub(r'[^a-zA-Z0-9_]', '_', col)
            alias = re.sub(r'_+', '_', alias)  # Replace multiple underscores with single
            alias = alias.strip('_')  # Remove leading/trailing underscores
            
            if alias != col:
                aliases[col] = alias
                reverse_aliases[alias] = col
                print_info(f"Column alias: '{col}' -> '{alias}'")
    
    return aliases, reverse_aliases


def get_sisso_equation(param_cols):
    """
    Unified function to get SISSO equation from user with step-by-step input
    """
    
    while True:  # Main loop for re-entry if user wants to correct
        print_header("SISSO EQUATION INPUT")
        
        # Step 1: Print instruction
        print_info("INPUT SISSO GIVEN EQUATION")
        print_info("\nAvailable parameters:")
        for i, col in enumerate(param_cols, 1):
            print(f"  {Fore.MAGENTA}{i}.{Style.RESET_ALL} {col}")

        print("Enter the equaiton in 'c0 + c1*(formula)' format:")
        print("NOW ENTER THE VALUES FOR: c0, c1, and formula:")
        
        # Step 2: Ask for C0
        print(f"\n{Fore.YELLOW}Enter coefficient values:{Style.RESET_ALL}")
        c0 = get_valid_float("  c0 = ")
        
        # Step 3: Ask for C1
        c1 = get_valid_float("  c1 = ")
        
        # Ask for additional coefficients
        print_info("\nDo you have more coefficients? (c2, c3, etc.)")
        more_coeffs = input(f"  {Fore.YELLOW}Enter 'y' for yes, 'n' for no:{Style.RESET_ALL} ").strip().lower()
        
        coeff_values = {'c0': c0, 'c1': c1}
        coeff_num = 2
        
        if more_coeffs == 'y':
            while True:
                try:
                    val = input(f"  c{coeff_num} = (press Enter to skip): ").strip()
                    if val == '':
                        break
                    coeff_values[f'c{coeff_num}'] = float(val)
                    coeff_num += 1
                except ValueError:
                    print(f"{Fore.RED}Invalid number. Try again.{Style.RESET_ALL}")
        
        # Step 4: Print all available symbols/operators
        print(f"\n{Fore.YELLOW}═══════════════════════════════════════════════════════{Style.RESET_ALL}")
        print(f"{Fore.CYAN}AVAILABLE OPERATORS AND SYMBOLS:{Style.RESET_ALL}")
        print(f"{Fore.YELLOW}═══════════════════════════════════════════════════════{Style.RESET_ALL}")
        print(f"\n{Fore.GREEN}Basic Operators:{Style.RESET_ALL}")
        print(f"  +         Addition")
        print(f"  -         Subtraction")
        print(f"  *         Multiplication")
        print(f"  /         Division")
        print(f"  **        Exponentiation (e.g., x**2)")
        
        print(f"\n{Fore.GREEN}Mathematical Functions:{Style.RESET_ALL}")
        print(f"  exp(x)    Exponential (e^x)")
        print(f"  sqrt(x)   Square root")
        print(f"  cbrt(x)   Cube root")
        print(f"  log(x)    Natural logarithm (ln)")
        print(f"  log10(x)  Base-10 logarithm")
        print(f"  abs(x)    Absolute value")
        print(f"  |x|       Absolute value (alternative)")
        
        print(f"\n{Fore.GREEN}Trigonometric Functions:{Style.RESET_ALL}")
        print(f"  sin(x)    Sine")
        print(f"  cos(x)    Cosine")
        print(f"  tan(x)    Tangent")
        
        print(f"\n{Fore.GREEN}Parentheses:{Style.RESET_ALL}")
        print(f"  ( )       Grouping operations")
        
        print(f"\n{Fore.YELLOW}Example equations:{Style.RESET_ALL}")
        print(f"  • c0 + c1*{param_cols[0]}")
        print(f"  • c0 + c1*{param_cols[0]}^2")
        if len(param_cols) > 1:
            print(f"  • c0 + c1*{param_cols[0]} + c2*{param_cols[1]}")
        if len(param_cols) > 2:
            print(f"  • c0 + c1*exp({param_cols[0]}) + c2*sqrt({param_cols[2]})")
            print(f"  • c0 + c1*({param_cols[0]}/{param_cols[1]})^2")
        print(f"{Fore.YELLOW}═══════════════════════════════════════════════════════{Style.RESET_ALL}\n")
        
        # Step 5: Ask user to enter the equation
        print_info("Enter the equation using coefficients and parameters:")
        print(f"{Fore.MAGENTA}Note: Use the coefficient names (c0, c1, c2, etc.) in your equation{Style.RESET_ALL}")
        equation_input = get_valid_string("\nEnter equation: ").strip()
        
        # Step 6: Combine coefficients and equation
        # The equation already has coefficient names, so we just need to store it
        combined_equation = equation_input
        
        # Display the complete equation with actual values
        display_equation = combined_equation
        for coeff_name, coeff_val in coeff_values.items():
            # Show what each coefficient represents
            display_equation = display_equation.replace(coeff_name, f"{coeff_val}")
        
        # Step 7: Ask if correct or want to re-enter
        print(f"\n{Fore.YELLOW}═══════════════════════════════════════════════════════{Style.RESET_ALL}")
        print_success("EQUATION SUMMARY:")
        print(f"\n{Fore.MAGENTA}Coefficients:{Style.RESET_ALL}")
        for coeff_name, coeff_val in sorted(coeff_values.items()):
            print(f"  {coeff_name} = {coeff_val}")
        
        print(f"\n{Fore.MAGENTA}Equation (symbolic):{Style.RESET_ALL}")
        print(f"  {combined_equation}")
        
        print(f"\n{Fore.CYAN}Equation (with values):{Style.RESET_ALL}")
        print(f"  {display_equation}")
        print(f"{Fore.YELLOW}═══════════════════════════════════════════════════════{Style.RESET_ALL}\n")
        
        confirm = input(f"{Fore.YELLOW}Is this correct? (y/n): {Style.RESET_ALL}").strip().lower()
        
        if confirm == 'y' or 'yes' or 'Y' or 'Yes' or 'YES':
            print_success("Equation confirmed!")
            break
        else:
            print_info("Let's re-enter the equation...\n")
            continue
    
    # Step 8: The rest - Process the equation
    print_info("\nProcessing equation...")
    
    # Preprocess equation to convert mathematical operators
    processed_equation, original_equation = preprocess_equation(combined_equation)
    
    print_info("Equation preprocessing complete:")
    print(f"  Original: {original_equation}")
    print(f"  Processed: {processed_equation}")
    
    # Extract coefficient names from the equation
    coeff_pattern = r'\b[cC]\d+\b'
    coeffs_in_equation = sorted(set(re.findall(coeff_pattern, combined_equation)))
    
    # Verify all coefficients in equation have values
    missing_coeffs = [c for c in coeffs_in_equation if c.lower() not in coeff_values]
    if missing_coeffs:
        print(f"{Fore.RED}Warning: Missing values for coefficients: {', '.join(missing_coeffs)}{Style.RESET_ALL}")
        for coeff in missing_coeffs:
            coeff_values[coeff.lower()] = get_valid_float(f"  {coeff} = ")
    
    print_success("All processing complete!")
    
    return combined_equation, coeff_values, processed_equation, display_equation


def preprocess_equation(equation):
    """
    Preprocess equation to convert mathematical operators to Python syntax
    Handles: ^, exp, sqrt, cbrt, log, abs, sin, cos, tan, etc.
    """
    # Store original for display
    original = equation
    
    # Replace ^ with ** for exponentiation (but not in exp^ patterns)
    equation = re.sub(r'exp\s*\^\s*\(', 'EXP_POWER(', equation)
    equation = equation.replace('^', '**')
    equation = equation.replace('EXP_POWER(', 'np.exp(')
    
    # Handle exp patterns
    equation = re.sub(r'exp-\s*\(', 'np.exp(-(', equation)
    equation = re.sub(r'\bexp\s*\(', 'np.exp(', equation)
    
    # Handle mathematical functions
    equation = re.sub(r'\bsqrt\s*\(', 'np.sqrt(', equation)
    equation = re.sub(r'\bcbrt\s*\(', 'np.cbrt(', equation)
    equation = re.sub(r'\blog\s*\(', 'np.log(', equation)
    equation = re.sub(r'\blog10\s*\(', 'np.log10(', equation)
    equation = re.sub(r'\bsin\s*\(', 'np.sin(', equation)
    equation = re.sub(r'\bcos\s*\(', 'np.cos(', equation)
    equation = re.sub(r'\btan\s*\(', 'np.tan(', equation)
    
    # Handle absolute value
    equation = re.sub(r'\|\-([^|]+)\-\|', r'np.abs(\1)', equation)
    equation = re.sub(r'\|([^|]+)\|', r'np.abs(\1)', equation)
    
    return equation, original


def apply_column_aliases(equation, aliases):
    """
    Replace column names in equation with their aliases
    """
    if not aliases:
        return equation
    
    # Sort by length (longest first) to avoid partial replacements
    sorted_cols = sorted(aliases.keys(), key=len, reverse=True)
    
    modified_equation = equation
    for col in sorted_cols:
        alias = aliases[col]
        modified_equation = re.sub(
            r'\b' + re.escape(col) + r'\b',
            alias,
            modified_equation
        )
    
    if modified_equation != equation:
        print_info("Equation with aliases:")
        print(f"  {modified_equation}")
    
    return modified_equation


def test_equation(df, param_cols, equation, coeff_values, n_samples=3, equation_display=None):
    """
    Test the equation on a few sample rows to verify it works
    """
    print_header("EQUATION VALIDATION")
    
    if equation_display:
        print_info(f"Testing equation: {equation_display}")
    
    print_info(f"Running test on {n_samples} sample rows...")
    
    try:
        # Get first n rows
        test_df = df[param_cols].head(n_samples)
        
        # Create namespace
        namespace = {}
        for col in param_cols:
            namespace[col] = test_df[col].values
        if coeff_values:
            namespace.update(coeff_values)
        namespace['np'] = np
        namespace['exp'] = np.exp
        namespace['sqrt'] = np.sqrt
        namespace['log'] = np.log
        namespace['log10'] = np.log10
        namespace['sin'] = np.sin
        namespace['cos'] = np.cos
        namespace['tan'] = np.tan
        namespace['abs'] = np.abs
        try:
            namespace['cbrt'] = np.cbrt
        except AttributeError:
            namespace['cbrt'] = lambda x: np.power(x, 1/3)
        
        # Evaluate
        results = eval(equation, {"__builtins__": {}}, namespace)
        
        # Display results
        print_success("Equation test passed!")
        print(f"\n{Fore.CYAN}{Style.BRIGHT}Sample Calculations:{Style.RESET_ALL}")
        
        for i in range(min(n_samples, len(results))):
            print(f"\n  {Fore.YELLOW}Sample {i+1}:{Style.RESET_ALL}")
            print(f"  {'-'*50}")
            for col in param_cols:
                val = test_df[col].iloc[i]
                print(f"    {col:30s} = {val:12.6e}")
            print(f"  {'-'*50}")
            print(f"    {Fore.GREEN}{'Predicted Value':30s} = {results[i]:12.6e}{Style.RESET_ALL}")
        
        return True
        
    except Exception as e:
        print_error(f"Equation test failed: {e}")
        print_warning("Check your equation syntax and parameter names")
        
        import traceback
        print(f"\n{Fore.RED}Error details:{Style.RESET_ALL}")
        traceback.print_exc()
        
        return False
    """Compute SISSO predictions based on equation"""
    print_header("COMPUTING SISSO PREDICTIONS")
    
    try:
        # Create a namespace with parameter columns and coefficients
        namespace = {}
        
        # Add parameter columns to namespace
        for col in param_cols:
            namespace[col] = df[col].values
            print_info(f"Loaded column: {col} (shape: {df[col].values.shape})")
        
        # Add coefficient values
        if coeff_values:
            namespace.update(coeff_values)
            print_info(f"Loaded {len(coeff_values)} coefficients")
        
        # Add numpy for mathematical functions
        namespace['np'] = np
        
        # Add some additional common math functions directly
        namespace['exp'] = np.exp
        namespace['sqrt'] = np.sqrt
        namespace['log'] = np.log
        namespace['log10'] = np.log10
        namespace['sin'] = np.sin
        namespace['cos'] = np.cos
        namespace['tan'] = np.tan
        namespace['abs'] = np.abs
        
        # Try to add cbrt if available (numpy >= 1.10)
        try:
            namespace['cbrt'] = np.cbrt
        except AttributeError:
            # Fallback for older numpy versions
            namespace['cbrt'] = lambda x: np.power(x, 1/3)
        
        print_info("Evaluating equation...")
        
        # Evaluate the equation
        y_pred = eval(equation, {"__builtins__": {}}, namespace)
        
        # Ensure output is a numpy array
        if not isinstance(y_pred, np.ndarray):
            y_pred = np.array(y_pred)
        
        print_success(f"Successfully computed {len(y_pred)} predictions")
        print_info(f"Prediction range: [{np.min(y_pred):.6e}, {np.max(y_pred):.6e}]")
        
        return y_pred
    
    except NameError as e:
        print_error(f"Variable name error: {e}")
        print_warning("Check that all column names in the equation match your selected parameters")
        print_info("Available parameters: " + ", ".join(param_cols))
        return None
    
    except SyntaxError as e:
        print_error(f"Equation syntax error: {e}")
        print_warning("Check your equation syntax. Common issues:")
        print("  - Missing operators (use * for multiplication)")
        print("  - Unmatched parentheses")
        print("  - Invalid function names")
        return None
    
    except ZeroDivisionError as e:
        print_error(f"Division by zero detected: {e}")
        print_warning("Some parameter values may be zero, causing division errors")
        return None
    
    except Exception as e:
        print_error(f"Error computing predictions: {e}")
        print_warning("Detailed error information:")
        print(f"  Equation: {equation}")
        print(f"  Error type: {type(e).__name__}")
        import traceback
        print(f"\n{Fore.RED}Traceback:{Style.RESET_ALL}")
        traceback.print_exc()
        return None

def compute_sisso_prediction(df, param_cols, equation, coeff_values):
    """
    Compute SISSO predictions based on equation
    
    Parameters:
    -----------
    df : DataFrame
        DataFrame containing parameter columns
    param_cols : list
        List of parameter column names
    equation : str
        SISSO equation (preprocessed with np. functions)
    coeff_values : dict
        Dictionary of coefficient values {c0: val, c1: val, ...}
    
    Returns:
    --------
    y_pred : numpy array
        Array of predicted values
    """
    print_header("COMPUTING SISSO PREDICTIONS")
    
    try:
        # Create a namespace with parameter columns and coefficients
        namespace = {}
        
        # Add parameter columns to namespace
        for col in param_cols:
            namespace[col] = df[col].values
            print_info(f"Loaded column: {col} (shape: {df[col].values.shape})")
        
        # Add coefficient values
        if coeff_values:
            namespace.update(coeff_values)
            print_info(f"Loaded {len(coeff_values)} coefficients")
        
        # Add numpy for mathematical functions
        namespace['np'] = np
        
        # Add some additional common math functions directly
        namespace['exp'] = np.exp
        namespace['sqrt'] = np.sqrt
        namespace['log'] = np.log
        namespace['log10'] = np.log10
        namespace['sin'] = np.sin
        namespace['cos'] = np.cos
        namespace['tan'] = np.tan
        namespace['abs'] = np.abs
        
        # Try to add cbrt if available (numpy >= 1.10)
        try:
            namespace['cbrt'] = np.cbrt
        except AttributeError:
            # Fallback for older numpy versions
            namespace['cbrt'] = lambda x: np.power(x, 1/3)
        
        print_info("Evaluating equation...")
        print_info(f"Equation: {equation}")
        
        # Evaluate the equation
        y_pred = eval(equation, {"__builtins__": {}}, namespace)
        
        # Ensure output is a numpy array
        if not isinstance(y_pred, np.ndarray):
            y_pred = np.array(y_pred)
        
        print_success(f"Successfully computed {len(y_pred)} predictions")
        print_info(f"Prediction range: [{np.min(y_pred):.6e}, {np.max(y_pred):.6e}]")
        print_info(f"Prediction mean: {np.mean(y_pred):.6e}")
        print_info(f"Prediction std: {np.std(y_pred):.6e}")
        
        return y_pred
    
    except NameError as e:
        print_error(f"Variable name error: {e}")
        print_warning("Check that all column names in the equation match your selected parameters")
        print_info("Available parameters: " + ", ".join(param_cols))
        return None
    
    except SyntaxError as e:
        print_error(f"Equation syntax error: {e}")
        print_warning("Check your equation syntax. Common issues:")
        print("  - Missing operators (use * for multiplication)")
        print("  - Unmatched parentheses")
        print("  - Invalid function names")
        return None
    
    except ZeroDivisionError as e:
        print_error(f"Division by zero detected: {e}")
        print_warning("Some parameter values may be zero, causing division errors")
        return None
    
    except Exception as e:
        print_error(f"Error computing predictions: {e}")
        print_warning("Detailed error information:")
        print(f"  Equation: {equation}")
        print(f"  Error type: {type(e).__name__}")
        import traceback
        print(f"\n{Fore.RED}Traceback:{Style.RESET_ALL}")
        traceback.print_exc()
        return None

def calculate_metrics(y_actual, y_pred, formulas):
    """Calculate evaluation metrics"""
    r2 = r2_score(y_actual, y_pred)
    rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
    mae = mean_absolute_error(y_actual, y_pred)
    errors = np.abs(y_actual - y_pred)
    max_error = np.max(errors)
    max_error_idx = np.argmax(errors)
    max_error_formula = formulas[max_error_idx]
    
    return {
        'r2': r2,
        'rmse': rmse,
        'mae': mae,
        'max_error': max_error,
        'max_error_formula': max_error_formula,
        'max_error_idx': max_error_idx,
        'errors': errors
    }


def auto_determine_axis_limits(y_actual, y_pred, padding=0.05):
    """Automatically determine appropriate axis limits with padding"""
    all_vals = np.concatenate([y_actual, y_pred])
    min_val = all_vals.min()
    max_val = all_vals.max()
    
    # Add padding
    data_range = max_val - min_val
    min_val -= data_range * padding
    max_val += data_range * padding
    
    return min_val, max_val


def create_plot(y_actual, y_pred, formulas, metrics, property_name, save_path):
    """Create and save the prediction plot"""
    print_header("CREATING PLOT")
    
    fig, ax = plt.subplots(figsize=(5.6, 5))
    
    # Main scatter plot
    ax.scatter(y_actual, y_pred, s=80, marker='o', color='crimson', 
               alpha=0.9, edgecolors='k', label='SISSO Prediction')
    
    # Annotation with metrics
    annotation_text = (
        f"$R^2$ = {metrics['r2']:.4f}\n"
        f"MAE = {metrics['mae']:.3e}\n"
        f"RMSE = {metrics['rmse']:.3e}\n"
        f"Max Error: {metrics['max_error']:.3e}\n"
        f"({metrics['max_error_formula']})"
    )
    
    ax.annotate(
        annotation_text,
        xy=(0.045, 0.941), xycoords='axes fraction',
        fontsize=12, ha='left', va='top',
        bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.8)
    )
    
    # Highlight max error point
    ax.scatter(
        y_actual[metrics['max_error_idx']],
        y_pred[metrics['max_error_idx']],
        color='blue', edgecolors='black', s=170, marker='x', 
        label='Max Error', linewidths=3
    )
    
    # Reference line (perfect prediction)
    min_val, max_val = auto_determine_axis_limits(y_actual, y_pred)
    ax.plot([min_val, max_val], [min_val, max_val], 'k--', lw=1)
    
    # Axis labels and formatting
    ax.set_xlabel(f"Actual {property_name}", fontsize=24)
    ax.set_ylabel(f"Predicted {property_name}", fontsize=24)
    ax.set_xlim(min_val, max_val)
    ax.set_ylim(min_val, max_val)
    
    # Tick formatting
    ax.xaxis.set_major_locator(MaxNLocator(nbins=6))
    ax.yaxis.set_major_locator(MaxNLocator(nbins=6))
    ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    ax.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))
    
    # Tick parameters
    ax.tick_params(axis='both', which='major', direction='in', labelsize=17, 
                   length=8, left=True, right=True, top=True, bottom=True)
    ax.tick_params(axis='both', which='minor', direction='in', length=3, width=1.0, 
                   left=True, right=True, top=True, bottom=True)
    ax.minorticks_on()
    ax.xaxis.set_minor_locator(AutoMinorLocator(5))
    ax.yaxis.set_minor_locator(AutoMinorLocator(5))
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=800, bbox_inches='tight')
    print_success(f"Plot saved: {save_path}")
    
    plt.show()


def save_results(df, formula_col, actual_col, y_pred, metrics, property_name, output_path, equation_original=None, display_equation=None):
    """Save prediction results to Excel"""
    print_header("SAVING RESULTS")
    
    # Create results DataFrame with all original columns plus predictions
    results_df = df.copy()
    results_df['SISSO_Prediction'] = y_pred
    results_df['Absolute_Error'] = metrics['errors']
    results_df['Relative_Error_%'] = (metrics['errors'] / np.abs(results_df[actual_col].values)) * 100
    
    # Create summary statistics
    summary_data = {
        'Metric': ['Property', 'Equation', 'Number of Data Points', 'R²', 'RMSE', 'MAE', 
                   'Max Error', 'Max Error Formula', 'Mean Absolute Error', 'Std of Errors'],
        'Value': [
            property_name,
            display_equation if display_equation else 'N/A',
            len(df),
            metrics['r2'],
            metrics['rmse'],
            metrics['mae'],
            metrics['max_error'],
            metrics['max_error_formula'],
            np.mean(metrics['errors']),
            np.std(metrics['errors'])
        ]
    }
    summary_df = pd.DataFrame(summary_data)
    
    # Save to Excel with multiple sheets
    with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
        results_df.to_excel(writer, sheet_name='Predictions', index=False)
        summary_df.to_excel(writer, sheet_name='Summary', index=False)
    
    print_success(f"Results saved: {output_path}")
    
    # Print summary to console
    print(f"\n{Fore.GREEN}{Style.BRIGHT}{'='*50}{Style.RESET_ALL}")
    print(f"{Fore.GREEN}{Style.BRIGHT}PREDICTION SUMMARY{Style.RESET_ALL}")
    print(f"{Fore.GREEN}{Style.BRIGHT}{'='*50}{Style.RESET_ALL}")
    print(f"R² Score:          {metrics['r2']:.6f}")
    print(f"RMSE:              {metrics['rmse']:.6e}")
    print(f"MAE:               {metrics['mae']:.6e}")
    print(f"Max Error:         {metrics['max_error']:.6e}")
    print(f"Max Error Formula: {metrics['max_error_formula']}")
    print(f"{Fore.GREEN}{Style.BRIGHT}{'='*50}{Style.RESET_ALL}\n")


def main():
    """Main execution function"""
    print(f"\n{Back.MAGENTA}{Fore.BLACK}{Style.BRIGHT}")
    print("="*60)
    print("    GENERALIZED SISSO PREDICTION TOOL")
    print("="*60)
    print(Style.RESET_ALL)
    
    # Step 1: Load data
    data_result = load_data()
    if not data_result:
        print_error("Failed to load data. Exiting.")
        return
    
    excel_path, sheet_name, df = data_result
    
    # Step 2: Select columns
    formula_col, actual_col, param_cols = select_columns(df)
    
    # Step 3: Get property name
    property_name = get_valid_string("\nEnter property name (for labeling): ")
    
    # Step 4: Get SISSO equation and coefficients
    equation, coeff_values, equation_original, display_equation = get_sisso_equation(param_cols)
    
    # Step 5: Prepare data
    print_header("PREPARING DATA")
    df_clean = df[[formula_col, actual_col] + param_cols].copy()
    df_clean.dropna(inplace=True)
    print_success(f"Cleaned data: {len(df_clean)} valid rows")
    
    # Step 5.5: Test equation on sample data
    if not test_equation(df_clean, param_cols, equation, coeff_values, n_samples=3, equation_display=equation_original):
        retry = input(f"\n{Fore.YELLOW}Equation test failed. Do you want to re-enter the equation? (y/n): {Style.RESET_ALL}").lower()
        if retry == 'y':
            equation, coeff_values, equation_original, display_equation = get_sisso_equation(param_cols)
            # Test again
            if not test_equation(df_clean, param_cols, equation, coeff_values, n_samples=3, equation_display=equation_original):
                print_error("Equation still failing. Exiting.")
                return
        else:
            print_warning("Proceeding with potentially incorrect equation...")
    
    proceed = input(f"\n{Fore.YELLOW}Proceed with full calculation? (y/n): {Style.RESET_ALL}").lower()
    if proceed == 'y' or 'yes' or 'Y' or 'Yes' or 'YES':
        print_info("Continuing...")
    else:
        print_info("Calculation cancelled by user.")
        return
    
    # Step 6: Compute predictions
    y_pred = compute_sisso_prediction(df_clean, param_cols, equation, coeff_values)
    if y_pred is None:
        print_error("Failed to compute predictions. Exiting.")
        return
    
    y_actual = df_clean[actual_col].values
    formulas = df_clean[formula_col].values
    
    # Step 7: Calculate metrics
    print_header("CALCULATING METRICS")
    metrics = calculate_metrics(y_actual, y_pred, formulas)
    print_success("Metrics calculated successfully")
    
    # Step 8: Create output folder
    print_header("SETTING UP OUTPUT")
    
    print_info("Step 1: Specify output folder location")
    default_output = str("./")
    print_info(f"Default output folder: {default_output}")
    
    use_default = input(f"{Fore.YELLOW}Use default folder? (y/n): {Style.RESET_ALL}").lower()
    
    if use_default == 'y':
        output_folder = default_output
    else:
        output_folder = get_valid_path("Enter output folder path: ", must_exist=False)
    
    # Optional subfolder
    create_subfolder = input(f"{Fore.YELLOW}Create a subfolder for this property? (y/n): {Style.RESET_ALL}").lower()
    
    if create_subfolder == 'y':
        subfolder_name = get_valid_string("Enter subfolder name: ")
        output_folder = str(Path(output_folder) / subfolder_name)
    
    # Create output folder
    Path(output_folder).mkdir(parents=True, exist_ok=True)
    print_success(f"Output folder ready: {output_folder}")
    
    # Step 9: Get file names
    print_info("\nStep 2: Specify output file names")
    
    # Suggest default names based on property
    default_plot = f"{property_name}_prediction_plot"
    default_results = f"{property_name}_SISSO_results"
    
    print_info(f"Default plot name: {default_plot}")
    use_default_plot = input(f"{Fore.YELLOW}Use default plot name? (y/n): {Style.RESET_ALL}").lower()
    
    if use_default_plot == 'y':
        plot_name = default_plot
    else:
        plot_name = get_valid_string("Enter plot filename (without extension): ")
    
    print_info(f"Default results name: {default_results}")
    use_default_results = input(f"{Fore.YELLOW}Use default results name? (y/n): {Style.RESET_ALL}").lower()
    
    if use_default_results == 'y':
        results_name = default_results
    else:
        results_name = get_valid_string("Enter results Excel filename (without extension): ")
    
    # Create full paths
    plot_path = Path(output_folder) / f"{plot_name}.pdf"
    results_path = Path(output_folder) / f"{results_name}.xlsx"
    
    # Step 10: Create plot
    create_plot(y_actual, y_pred, formulas, metrics, property_name, str(plot_path))
    
    # Step 11: Save results
    save_results(df_clean, formula_col, actual_col, y_pred, metrics, property_name, str(results_path), equation_original, display_equation)
    
    # Final message
    print(f"\n{Back.GREEN}{Fore.BLACK}{Style.BRIGHT}")
    print("="*60)
    print("    ANALYSIS COMPLETE!")
    print("="*60)
    print(Style.RESET_ALL)
    
    print(f"\n{Fore.CYAN}{Style.BRIGHT}FILE SUMMARY:{Style.RESET_ALL}")
    print(f"{Fore.YELLOW}Input:{Style.RESET_ALL}")
    print(f"  File: {excel_path}")
    print(f"  Sheet: {sheet_name}")
    print(f"\n{Fore.YELLOW}Output:{Style.RESET_ALL}")
    print(f"  Plot: {plot_path}")
    print(f"  Results: {results_path}")


if __name__ == "__main__":
    try:
        main()
    except KeyboardInterrupt:
        print(f"\n\n{Fore.YELLOW}Operation cancelled by user.{Style.RESET_ALL}")
    except Exception as e:
        print_error(f"Unexpected error: {e}")
        import traceback
        traceback.print_exc()
```
